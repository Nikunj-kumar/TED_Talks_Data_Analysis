```python
def ted_str(ratings_str):
    return ast.literal_eval(ratings_str)
```

```python
ted_str(ted.ratings[0])
```

```python
#Using lamda function for extracting rating column
ted['rating_list'] = ted.ratings.apply(lambda x: ast.literal_eval(x))
```

# Count the total number of rating recieved by each talk
    * Make a new column for rating recieved "num_ratings"
  
  Extra
      
      1.For each talk calculate the number of ratings that were negative.
      
      2.For each talk, calculate the average number of ratings it recieved per day, since it's release day
      

```python
# From first rating_list , we will extract the dict one by one
ted.rating_list[0][0]
```

```python
# We will make a function for above query or we can also sum it by dataframe function
for i in range(2550):
    print(pd.DataFrame(ted.rating_list[i])['count'].sum())
```

```python
#Function to sum up whole rating recieved for each talk 
def get_num_rating(list_of_dict):
    num=0
    for d in list_of_dict:
        num = num + d['count']
    return num
```

```python
get_num_rating(ted.rating_list[0])
```

```python
#Using apply method to applt the above function on whole rating data
ted.rating_list.apply(get_num_rating)
```

```python
#Assigning a new column "num/-rating"
ted['num_rating'] = ted.rating_list.apply(get_num_rating)
```

# Which occupations deliver the funniest TED talks on average

```python
# Count the number of funny rating
for i in range(14):
    print(ted.rating_list[0][i]['name']=='Funny')
```

```python
#Defining function to count the number of funny talk
def funny_rating(list_of_dict):
    for d in list_of_dict:
        if d['name'] == 'Funny':
            return d['count']
```

```python
funny_rating(ted.rating_list[1])
```

```python
ted['funny_ratings'] = ted.rating_list.apply(funny_rating)
```

```python
#Percentage of funny talk from overall talk review recieved for each talk.
ted['Fun_per'] = ted['funny_ratings'] / ted['num_rating'] * 100
```

```python
ted['Fun_per']
```

```python
#Checking whether above classification make any sense or not
ted.sort_values('Fun_per').speaker_occupation.tail(20)
```

```python
## Analyse funny rate percentage by occupation
ted.groupby('speaker_occupation').Fun_per.mean().sort_values().tail()
```
